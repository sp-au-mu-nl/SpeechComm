```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
yA, sr = librosa.load('a8k.wav',sr=None)
tA = np.arange(0,len(yA))/sr
px.line(x=tA,y=yA)
```

```python
fig = px.line(x=tA,y=yA)
fig.update_yaxes({"tickfont_size": 16})
fig.update_xaxes({"tickfont_size": 16})
fig.write_image('fig01_07.pdf')
```

```python
fig = px.line(x=tA,y=yA)
fig.update_yaxes({"tickfont_size": 16})
fig.update_xaxes({"tickfont_size": 16,
                  "range": [0.3,0.4]})
fig.write_image('fig01_08.pdf')
```

```python
plt.figure()
yi, fs = librosa.load('i8k.wav', sr=None)
ti = np.arange(0,len(yi))/fs
idx = np.logical_and(ti>0.327,ti < 0.346)
plt.plot(ti[idx],yi[idx])
```
